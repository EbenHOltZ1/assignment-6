```python
import os
import sys
import numpy as np
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as patches
import matplotlib.pyplot as plt
```

```python
#create an array
test_data = np.array([[5,7,9,2,40],[4,2,6,8,13],[10,2,5,11,6],[2,5,6,8,1],[10,4,8,3,6]])
print(test_data)
print(test_data[0])
print(test_data[0][0])
```

```python
#Subset the test_data
sub_data = test_data[0][(test_data[0] < 8) & (test_data[0] > 3)]
print(sub_data)
```

```python
#Get the indices of the elements
idx = np.where(np.isin(test_data[0],sub_data))[0]
also_idx = np.flatnonzero(np.isin(test_data[0],sub_data))
print(idx)
print(also_idx)
```

```python
#Create a dataframe
test_pd = {
    "a":pd.Series(["frame1","frame2"]),
    "b":pd.Series([2,4])
}
test_df = pd.DataFrame(test_pd)
print(test_df)
```

```python
print(test_df["a"])
print(test_df["a"][0])
print(test_df.loc[0,"a"])
```

```python
#Create heatmap
#Generate colormap, without vmax setting
cm = LinearSegmentedColormap.from_list('custom color',[(0,'white'),(0.5,'orange'),(1,'red')])
plt.imshow(test_data,cmap=cm)
#show title
plt.title("Your title")
#show colorbar
plt.colorbar()

plt.show()
```

```python
#Create heatmap
#Generate colormap
cm = LinearSegmentedColormap.from_list('custom color',[(0,'white'),(0.5,'orange'),(1,'red')])
#specify the max value here
plt.imshow(test_data,cmap=cm,vmin=0,vmax=np.percentile(test_data,95))
#show title
plt.title("Your title")
#show colorbar
plt.colorbar()

plt.show()
```

```python
##Add rectangle to the plot using patches.Rectangle
#Create heatmap
#Generate colormap
cm = LinearSegmentedColormap.from_list('custom color',[(0,'white'),(0.5,'orange'),(1,'red')])
#patches.Rectangle(anchor_point(x,y),width,height,...)
rec = patches.Rectangle((2,1),2,1,linewidth=1,linestyle="dashed",alpha=0.5,edgecolor='b', facecolor='none')
#specify the max value here
plt.imshow(test_data,cmap=cm,vmin=0,vmax=np.percentile(test_data,95))
# Get Current Reference
ax = plt.gca()
#Add the rectangle to thte plot
ax.add_patch(rec)
#show title
plt.title("Your title")
#show colorbar
plt.colorbar()

plt.show()
```
